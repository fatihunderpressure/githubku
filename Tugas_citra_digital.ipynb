**Konversi RGB ke grayscale**

**Sesuai**

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/content/Screenshot 2026-02-15 141743.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.axis('off')
plt.show()
```

**Grayscale** **ke biner**

**Sesuai**

```python
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(binary, cmap='gray')
plt.axis('off')
plt.show()
```

**Grayscale ke M-BIT**

**Sesuai**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/Screenshot 2026-02-15 141743.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

mbit = (gray // 64) * 64

plt.imshow(mbit, cmap='gray')
plt.axis('off')
plt.show()
```

**Brightness**

```python
bright = cv2.convertScaleAbs(img, beta=50)

plt.imshow(cv2.cvtColor(bright, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

**Contrast**

```python
contrast = cv2.convertScaleAbs(img, alpha=2, beta=0)
contrast = cv2.cvtColor(contrast, cv2.COLOR_BGR2RGB)

plt.imshow(contrast)
plt.axis('off')
plt.show()
```

**Grayscale ke Histogram**
