```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# membaca gambar
img = cv2.imread('bahlil.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 1. Negatif
negatif = 255 - img_rgb

# 2. Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 3. Blur
blur = cv2.GaussianBlur(img_rgb,(15,15),0)

# 4. Edge Detection
edges = cv2.Canny(img,100,200)

# 5. Sharpen
kernel = np.array([[0,-1,0],
                   [-1,5,-1],
                   [0,-1,0]])
sharpen = cv2.filter2D(img_rgb,-1,kernel)

# menampilkan semua gambar
plt.figure(figsize=(10,8))

plt.subplot(2,3,1)
plt.imshow(img_rgb)
plt.title("Gambar Asli")
plt.axis("off")

plt.subplot(2,3,2)
plt.imshow(negatif)
plt.title("Negatif")
plt.axis("off")

plt.subplot(2,3,3)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale")
plt.axis("off")

plt.subplot(2,3,4)
plt.imshow(blur)
plt.title("Blur")
plt.axis("off")

plt.subplot(2,3,5)
plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.axis("off")

plt.subplot(2,3,6)
plt.imshow(sharpen)
plt.title("Sharpen")
plt.axis("off")

plt.show()
```
